```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,30,40,50]
y1=[21,45,34,32,12]
y2=[12,23,32,14,45]

plt.figure(figsize=[9,5])
plt.plot(x,y1,label="Female")
plt.plot(x,y,label="Male")
plt.plot(x,y2,label="Kids")
plt.legend(["a1","a2","a3"],loc="best",ncols=2) #loc=location
plt.show()
```
