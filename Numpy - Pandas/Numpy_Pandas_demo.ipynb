
#Merge, join and concatenate in pandas


```python
import pandas as pd
data1={"EmployeeID":["E01","E02","E03","E04"],
       "Names":["Ram","shyam","Rima","sima"]}
data2={"EmployeeID":["E01","E05","E03","E06"],
       "Salary":[100,200,300,400]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1)
print()
print(df2)
```

```python
print(pd.merge(df1,df2,on="EmployeeID"))
```

```python
print(pd.merge(left=df2,right=df1,on="EmployeeID",how="left"))
```

```python
print(pd.merge(left=df1,right=df2,on="EmployeeID",how="right"))
```


#compare dataframes


```python
import pandas as pd
dictionary = {"fruits":["mango","apple","banana","papaya"],
              "price":[100,200,150,50],
              "quantity":[15,10,30,33]}
df1=pd.DataFrame(dictionary)
print(df1)

df2=df1.copy()

df2.loc[0,"price"]=120
df2.loc[1,"price"]=220
df2.loc[2,"price"]=400
df2.loc[0,"quantity"]=20
df2.loc[1,"quantity"]=15
df2.loc[2,"quantity"]=35

print(df2)
print(df1.compare(df2,align_axis=0))
print(df1.compare(df2,keep_shape=True))
```


#Pivoting and Melting DataFrames - Pandas


```python
import pandas as pd
dictionary = {"Keys":["K1","K2","K1","K2"],
              "Names":["john","Ben","David","Peter"],
              "Houses":["Green","blue","red","yellow"],
              "Grades":["3rd","4th","5th","6th"]}
df=pd.DataFrame(dictionary)
print(df)
print(df.pivot(index="Keys",columns="Names",values=["Houses","Grades"]))
```

```python
import pandas as pd

dictionary = {"Names": ["john", "Ben", "David", "Peter"],
              "Houses": ["Green", "blue", "red", "yellow"],
              "Grades": ["3rd", "4th", "5th", "6th"]}
df = pd.DataFrame(dictionary)
print(df)
print(pd.melt(df, id_vars=["Names"],value_vars=["Houses","Grades"],var_name="Houses&Grades",value_name="value"))
# print(df.pivot(index="Keys", columns="Names", values=["Houses", "Grades"]))
```
