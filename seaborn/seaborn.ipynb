```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data=sns.load_dataset("tips")
print(data)

gp=data.groupby("day").agg({"tip":"count"})
print(gp)
sns.heatmap(gp)
plt.show()
```
