```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Sales-Export_2019-2020.csv")

# Step 1: Strip leading/trailing whitespace from column names
df.columns = df.columns.str.strip()

# Step 2: Remove duplicate rows
df = df.drop_duplicates()

# Step 3: Convert 'order_value_EUR' to numeric
df['order_value_EUR'] = df['order_value_EUR'].str.replace(',', '', regex=False)
df['order_value_EUR'] = pd.to_numeric(df['order_value_EUR'], errors='coerce')

# Step 4: Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])
df['date']=df['date'].dt.strftime('%d-%m-%y')

# Step 5: Standardize string columns (lowercase + strip)
for col in df.select_dtypes(include='object').columns:
    if col not in ['order_id']:  # Preserve order_id format
        df[col] = df[col].str.lower().str.strip()

# Step 6: Final check
print(df.info())
print(df.head())

# Step 7: Save cleaned dataset
df.to_csv('cleaned_sales_export.csv', index=False)
print(" Cleaned dataset saved as 'cleaned_sales_export.csv'")
```
